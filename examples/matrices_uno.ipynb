```python
import numpy as np

def generar_matrices(n_matrices, condicion="pares"):
    """
    Genera un número 'n' de matrices de 4x4 bajo una condición específica.
    """
    matrices = [] 
    
    for i in range(n_matrices):
        if condicion == "pares":
            matriz = np.random.randint(1, 51, size=(4, 4)) * 2
        elif condicion == "impares":
            matriz = np.random.randint(0, 50, size=(4, 4)) * 2 + 1
        elif condicion == "multiplos_de_5":
            matriz = np.random.randint(1, 21, size=(4, 4)) * 5
        else:
            matriz = np.random.randint(1, 101, size=(4, 4))
            
        matrices.append(matriz)
        
    return matrices

#condiciones
cantidad = 8               
tipo_condicion = "pares"   

#se generan matrices
mis_matrices = generar_matrices(cantidad, tipo_condicion)

print(f"Generando {cantidad} matrices de 4x4 con la condición: {tipo_condicion}\n")

for i, matriz in enumerate(mis_matrices):
    print(f"--- Matriz {i + 1} ---")
    print(matriz)
    print("")
```
