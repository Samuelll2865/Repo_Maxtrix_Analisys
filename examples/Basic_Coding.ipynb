#### Manual and Automatic Matrix Construction

Matrices can be created manually by directly specifying their entries, or automatically by using loops and numerical functions.  
Both approaches are useful in engineering and scientific programming.

The manual definition is convenient for small examples and for checking results by hand, while the automatic construction becomes much more efficient when dealing with larger systems or repetitive patterns.

```python
a  =  np.sum(2+3)
print("a = ",a)
b  = 'hi'
print('b = ', b)
c = np.array([1,2,3,4])
print('c is', c.astype, 'which value is =', c)

#--- Basic loop for array construction
cc = np.zeros(4)
for i in np.arange(1, 4, 1):
    cc[i] = i+1
cc[0] = 1

#--- Basic loop for array construction
dd = []
for j in np.arange(1,4+1,1):
    dd.append(j)

print('cc = ', cc, ', dd = ', dd)

manu = np.array([[1,5,9,13],[2,6,10,14],[3,7,11,15],[4,8,12,16]])

#--- Basic loop for array construction "Matrix"
auto = np.zeros([4,4])
i = 0
for j in np.arange(1,4+1,1):
    auto[i,:] = np.linspace(j,j+12,4)
    i = i+1

print('\033[1;32m Manual Matrix Array : \033[0m\n', manu)
print('\033[1;36m Automatic Matrix Array : \033[0m\n', auto)
```
